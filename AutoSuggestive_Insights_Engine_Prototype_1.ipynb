```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

def get_suggestions(query):
    urlbase = 'https://www.google.com/search?q='
    url = urlbase + query + '&num=100'

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')  # Optional: Run Chrome in headless mode (no GUI)
    chrome_options.add_argument('--disable-extensions')
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--no-sandbox')

    driver = webdriver.Chrome(options=chrome_options)
    driver.get(url)

    time.sleep(5) # Wait for the page to load

    suggest_ele = driver.find_elements(By.CSS_SELECTOR, 'div')

    suggested_eles = [ele.text for ele in suggest_ele]

    driver.quit()

    return suggested_eles

def main():
    query = input('Enter your search query : ')
    suggestions = get_suggestions(query)

    for i, suggestion in enumerate(suggestions, start=1):
        print(f'{i}:{suggestion}')

if __name__ == '__main__':
    main()
```
